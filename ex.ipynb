```python
import time
import random
import string

def busca_ingenua(texto, padrao):
    n = len(texto)
    m = len(padrao)

    iteracoes = 0
    instrucoes = 0

    if m == 0:
        return 0, iteracoes, instrucoes

    for i in range(n - m + 1):
        instrucoes += 1

        j = 0
        instrucoes += 1

        while j < m:
            iteracoes += 1
            instrucoes += 1

            if texto[i + j] != padrao[j]:
                instrucoes += 1
                break

            j += 1
            instrucoes += 1

        if j == m:
            instrucoes += 1
            return i, iteracoes, instrucoes

    return -1, iteracoes, instrucoes

def hash_horner(s, m, R, Q):
    h = 0
    iteracoes = 0
    instrucoes = 0

    for j in range(m):
        iteracoes += 1
        h = (h * R + ord(s[j])) % Q
        instrucoes += 1

    return h, iteracoes, instrucoes


def rabin_karp(texto, padrao):
    n = len(texto)
    m = len(padrao)

    R = 256
    Q = 2147483647  # primo grande

    iteracoes = 0
    instrucoes = 0

    if m == 0:
        return 0, iteracoes, instrucoes

    if m > n:
        return -1, iteracoes, instrucoes

    hash_padrao, it, inst = hash_horner(padrao, m, R, Q)
    iteracoes += it
    instrucoes += inst

    for i in range(n - m + 1):
        iteracoes += 1

        trecho = texto[i:i + m]
        instrucoes += 1

        hash_texto, it, inst = hash_horner(trecho, m, R, Q)
        iteracoes += it
        instrucoes += inst

        if hash_padrao == hash_texto:
            instrucoes += 1

            if trecho == padrao:
                instrucoes += 1
                return i, iteracoes, instrucoes

    return -1, iteracoes, instrucoes


def calcula_lps(padrao):
    m = len(padrao)
    lps = [0] * m

    tamanho = 0
    i = 1

    iteracoes = 0
    instrucoes = 0

    while i < m:
        iteracoes += 1

        if padrao[i] == padrao[tamanho]:
            tamanho += 1
            lps[i] = tamanho
            i += 1
            instrucoes += 3
        else:
            if tamanho != 0:
                tamanho = lps[tamanho - 1]
                instrucoes += 1
            else:
                lps[i] = 0
                i += 1
                instrucoes += 2

    return lps, iteracoes, instrucoes


def kmp(texto, padrao):
    n = len(texto)
    m = len(padrao)

    iteracoes = 0
    instrucoes = 0

    if m == 0:
        return 0, iteracoes, instrucoes

    if m > n:
        return -1, iteracoes, instrucoes

    lps, it, inst = calcula_lps(padrao)
    iteracoes += it
    instrucoes += inst

    i = 0
    j = 0

    while i < n:
        iteracoes += 1

        if texto[i] == padrao[j]:
            i += 1
            j += 1
            instrucoes += 2

        if j == m:
            instrucoes += 1
            return i - j, iteracoes, instrucoes

        elif i < n and texto[i] != padrao[j]:
            instrucoes += 1

            if j != 0:
                j = lps[j - 1]
                instrucoes += 1
            else:
                i += 1
                instrucoes += 1

    return -1, iteracoes, instrucoes



def gerar_texto(tamanho):
    letras = string.ascii_uppercase
    return ''.join(random.choice(letras) for _ in range(tamanho))


def testar_algoritmo(nome, funcao, texto, padrao):
    inicio = time.time()

    posicao, iteracoes, instrucoes = funcao(texto, padrao)

    fim = time.time()
    tempo = fim - inicio

    print(f"\nAlgoritmo: {nome}")
    print(f"Posição encontrada: {posicao}")
    print(f"Iterações: {iteracoes}")
    print(f"Instruções aproximadas: {instrucoes}")
    print(f"Tempo: {tempo:.6f} segundos")



texto = "ABCDCBDCBDACBDABDCBADF"
padrao = "ADF"

testar_algoritmo("Busca Ingênua", busca_ingenua, texto, padrao)
testar_algoritmo("Rabin-Karp", rabin_karp, texto, padrao)
testar_algoritmo("KMP", kmp, texto, padrao)



texto_grande = gerar_texto(5000000)

# Colocando um padrão conhecido no final do texto
padrao_grande = "XYZABC"
texto_grande = texto_grande + padrao_grande

testar_algoritmo("Busca Ingênua - Grande", busca_ingenua, texto_grande, padrao_grande)
testar_algoritmo("Rabin-Karp - Grande", rabin_karp, texto_grande, padrao_grande)
testar_algoritmo("KMP - Grande", kmp, texto_grande, padrao_grande)
```
